```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("Scouting Data.csv", index_col="Team")
df.rename(columns=lambda x: x.replace("\n", " "), inplace=True)

#UNCOMMENT ONLY IF DATA DOES NOT HAVE NAN VALUES
# scaler = MinMaxScaler()
# vals = list(df)
# for i in vals:
#     if "%" in i:
#         for z in df[i]:
#             z.replace("%", "")
#     else:
#         pass
# df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

paramls = []
weightls = []
broken_vals = []
team_dict = {}
x = int(input("How many paramters would you like to sort by?: "))
y = input("Would you like to add your own weights? (Yes or No): ")

if y == "Yes":
    for i in range(x):
        paramls.append(input(f"Parameter #{i+1}: "))
        weightls.append(float(input(f"Weight for param #{i+1}")))
    for index, row in df.iterrows():
        param_avg = 0.0
        try:
            for z, x in zip(paramls, weightls):
                if "%" in z: 
                    param_avg += float(str(row[z]).replace("%", ""))*float(x)
                else:
                    param_avg += float(row[z])
            team_dict[index] = param_avg/len(paramls)
        except ValueError:
             team_dict[index] = float("-inf")
             broken_vals.append(index)
elif y == "No":
    for i in range(x):
        paramls.append(input(f"Parameter #{i+1}: "))
    for index, row in df.iterrows():
        param_avg = 0.0
        try:
            for z in paramls:
                if "%" in z: 
                    param_avg += float(str(row[z]).replace("%", ""))
                else:
                    param_avg += float(row[z])
            team_dict[index] = param_avg/len(paramls)
        except ValueError:
             team_dict[index] = float("-inf")
             broken_vals.append(index)
print(team_dict)
print(broken_vals)
sorted_d = dict(sorted(team_dict.items(), key=lambda x:x[1], reverse = True))
print(sorted_d)
index = list(sorted_d.keys())
print(index)

new_df = pd.DataFrame(columns=df.columns)
for i in index:
    new_df = new_df.append(df.loc[i])

def highlight_rows(row):
    if row.name in broken_vals:
        return ['background-color: red'] * len(row)
    else:
        return [''] * len(row)

styled_df = new_df.style.apply(highlight_rows, axis=1)
styled_df.to_excel("Final Data.xlsx")
```
